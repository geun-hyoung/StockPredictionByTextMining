### 주식 보조지표 추가

```python
# 해당 영업일의 거래 대금 = 거래량 * 종가 - 출처 링크 첨부
tsla_df['transaction_amount'] = tsla_df['Close'] * tsla_df['Volume']
tsla_df.head()
```

```python
# 해당 영업일의 거래 대금의 변동 비율(전일 대비)
tsla_df['transaction_ratio'] = tsla_df['transaction_amount'].pct_change()
tsla_df.head()
```

```python
# 해당 영업일의 종가 변동 비율(전일 대비)
tsla_df['Close_ratio'] = tsla_df['Close'].pct_change()
tsla_df.head()
```

```python
# 이동평균, window = 이동평균 주기
ma = [5,20,60,120]
for days in ma:
    tsla_df['ma_'+str(days)] = tsla_df['Close'].rolling(window = days).mean()

tsla_df.head()
```

```python
# 상대 강도 지수(RSI): 주식이 과매수 혹은 과매도 상태인지를 나타내는 지표
delta = tsla_df['Close'].diff()

up = delta.clip(lower=0)
down = -1*delta.clip(upper=0)

ema_up = up.ewm(com=13, adjust=False).mean()
ema_down = down.ewm(com=13, adjust=False).mean()

rs = ema_up/ema_down
tsla_df['RSI'] = 100 - (100/(1 + rs))
tsla_df.head()
```

```python
# 2019년 6월 부터 불러와서 ma_5, ma_6 등 값을 구해 놓기 위해서
tsla_df = tsla_df[tsla_df.index >= '2020-01-01']
tsla_df.head()
```

```python
# 주식 지표 베타(?)
returns = tsla_df['Close'].pct_change()
sp500 = yf.download('^GSPC', start='2019-01-01', end='2022-12-31')
sp500_returns = sp500['Close'].pct_change()
covariance = returns.cov(sp500_returns)
variance = sp500_returns.var()
beta = covariance / variance
print("Beta: ", beta)
```

## 기존 예측 모델 : LSTM(RNN) 등

```python
tsla_df.loc[:, 'Close_ratio'] *= 100
```

```python
tsla_df.describe()
```

```python
# Close_ratio 열의 값이 음수인 경우 0, 양수인 경우 1로 변환
tsla_df['Close_ratio'] = tsla_df['Close_ratio'].apply(lambda x: 0 if x < 3 else 1)
tsla_df
```

```python
scaler = MinMaxScaler()

scaled_df = scaler.fit_transform(tsla_df)
scaled_df = pd.DataFrame(scaled_df)

scale_cols = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume', 'transaction_amount', 'transaction_ratio','Close_ratio', 'ma_5', 'ma_20', 'ma_60', 'ma_120', 'RSI']
scaled_df.columns = scale_cols
scaled_df
```

```python
TEST_SIZE = 200 # TEST_SIZE 개수만큼의 데이터를 테스트 데이터로, 나머지를 훈련 데이터로 나눔
WINDOW_SIZE = 10

train = scaled_df[:-TEST_SIZE] #0~556
test = scaled_df[-TEST_SIZE:] #556~

train = train.drop(['Adj Close'], axis = 1)
test = test.drop(['Adj Close'], axis = 1)
train# print(test)
```

```python
def make_dataset(data, label, window_size=20):
    feature_list = []
    label_list = []
    for i in range(len(data) - window_size):
        feature_list.append(np.array(data.iloc[i:i+window_size]))
        label_list.append(np.array(label.iloc[i+window_size]))
    return np.array(feature_list), np.array(label_list)
```

```python
from sklearn.model_selection import train_test_split

feature_cols = ['Open', 'High', 'Low', 'Volume', 'transaction_amount', 'transaction_ratio', 'ma_5', 'ma_20', 'ma_60', 'ma_120', 'RSI']
label_cols = ['Close_ratio']

train_feature = train[feature_cols]
train_label = train[label_cols]

train_feature, train_label = make_dataset(train_feature, train_label, 20)

x_train, x_valid, y_train, y_valid = train_test_split(train_feature, train_label, test_size=0.2)
print(x_train.shape, x_valid.shape)
print(y_train.shape, y_valid.shape)
```

```python
test_feature = test[feature_cols]
test_label = test[label_cols]

test_feature.shape, test_label.shape
```

```python
test_feature, test_label = make_dataset(test_feature, test_label, 20)
test_feature.shape, test_label.shape
```

```python
model = Sequential()

model.add(LSTM(units=128, return_sequences=True, input_shape=(x_train.shape[1], x_train.shape[2])))
model.add(LSTM(units=128, return_sequences=True))
model.add(LSTM(units=128, return_sequences=True))

model.add(Dense(units=1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

history = model.fit(x_train, y_train, epochs=100, batch_size=32, validation_data=(x_valid, y_valid))

train_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(train_acc, label='Train Accuracy')
plt.title('Training Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(val_acc, label='Validation Accuracy', color='orange')
plt.title('Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```
